```python
import astropy
import numpy as np
import math
import scipy
from astropy.table import Table
import pandas
import re
import matplotlib.pyplot as plt
```

```python
all_cols = data.colnames

# Separate into ctrl and non-ctrl columns
ctrl_cols = [col for col in all_cols if 'ctrl' in col]
non_ctrl_cols = [col for col in all_cols if 'ctrl' not in col]

# Create two new tables
ctrl_table = data[ctrl_cols]
non_ctrl_table = data[non_ctrl_cols]

```

```python
t = data
all_cols = t.colnames

# Identify control columns
ctrl_cols = [col for col in all_cols if 'ctrl' in col]

# Group control columns by gene
from collections import defaultdict

gene_to_ctrls = defaultdict(list)

for col in ctrl_cols:
    match = re.match(r'g(\d+)\+ctrl(?:_\d+)?', col)
    if match:
        gene_id = match.group(1)
        gene_to_ctrls[gene_id].append(col)

# Identify pairwise (non-control) columns
pair_cols = [col for col in all_cols if 'ctrl' not in col]

# For each pair column, compute control sum
results = {}

for col in pair_cols:
    # Extract the two gene IDs
    match = re.match(r'g(\d+)\+g(\d+)', col)
    if not match:
        print(f"Skipping non-matching column: {col}")
        continue

    gene1, gene2 = match.group(1), match.group(2)

    # Find matching control columns
    gene1_ctrls = gene_to_ctrls.get(gene1, [])
    gene2_ctrls = gene_to_ctrls.get(gene2, [])

    if not gene1_ctrls or not gene2_ctrls:
        print(f"Missing ctrl columns for {col}: gene1={gene1_ctrls}, gene2={gene2_ctrls}")
        continue

    # Compute the average control values for each gene
    gene1_ctrl_avg = np.mean([t[ctrl] for ctrl in gene1_ctrls], axis=0)
    gene2_ctrl_avg = np.mean([t[ctrl] for ctrl in gene2_ctrls], axis=0)

    # Sum the two averages
    ctrl_sum = gene1_ctrl_avg + gene2_ctrl_avg
    ctrl_product = gene1_ctrl_avg * gene2_ctrl_avg

    # Store these new columns
    new_col_name = f"g{gene1} u g{gene2}"
    t[new_col_name] = ctrl_sum
    
    product_col_name = f"g{gene1} x g{gene2}"
    t[product_col_name] = ctrl_product

    observed = t[col]

    # Calculate Residuals
    sum_residual = ctrl_sum - observed
    product_residual = ctrl_product - observed
    combined_residual = (ctrl_sum + ctrl_product) - observed

    # Store residuals
    t[f"g{gene1} u_res g{gene2}"] = sum_residual
    t[f"g{gene1} x_res g{gene2}"] = product_residual
    t[f"g{gene1} ux_res g{gene2}"] = combined_residual
```

```python

row_idx = 0  # or any other row number

u_res_cols = [col for col in t.colnames if 'u_res' in col]
x_res_cols = [col for col in t.colnames if 'x_res' in col]
ux_res_cols = [col for col in t.colnames if 'ux_res' in col]

u_res_values = np.array([t[col][row_idx] for col in u_res_cols])
x_res_values = np.array([t[col][row_idx] for col in x_res_cols])
ux_res_values = np.array([t[col][row_idx] for col in ux_res_cols])

plt.figure(figsize=(8, 5))
plt.hist(u_res_values, bins=20, density=True, alpha=0.4, label='u')
plt.hist(x_res_values, bins=20, density=True, alpha=0.4, label='x')
plt.hist(ux_res_values, bins=20, density=True, alpha=0.4, label='ux')
plt.title(f'Residual Histogram for Row {row_idx}')
plt.xlabel('Residual Value')
plt.ylabel('Count')
# plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
u_std = np.std(u_res_values)
x_std = np.std(x_res_values)
ux_std = np.std(ux_res_values)

u_pull = u_res_values / u_std
x_pull = x_res_values / x_std
ux_pull = ux_res_values / ux_std
```

```python
rows = len(t)
pull_stats = Table(names=["row", "u_mean", "u_std", "x_mean", "x_std", "ux_mean", "ux_std"],
                   dtype=[int, float, float, float, float, float, float])

for row_idx in range(rows):
    u_res = np.array([t[col][row_idx] for col in u_res_cols])
    x_res = np.array([t[col][row_idx] for col in x_res_cols])
    ux_res = np.array([t[col][row_idx] for col in ux_res_cols])
    
    # Avoid division by zero
    u_std = np.std(u_res) if np.std(u_res) != 0 else 1
    x_std = np.std(x_res) if np.std(x_res) != 0 else 1
    ux_std = np.std(ux_res) if np.std(ux_res) != 0 else 1

    u_pull = u_res / u_std
    x_pull = x_res / x_std
    ux_pull = ux_res / ux_std

    pull_stats.add_row([
        row_idx,
        np.mean(u_pull), np.std(u_pull),
        np.mean(x_pull), np.std(x_pull),
        np.mean(ux_pull), np.std(ux_pull)
    ])
```
